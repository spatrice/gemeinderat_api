# Gemeinderat API


### Verbinde mit der Gemeinderat API um alle Geschäfte herunterzuladen.

```python
import requests
import pandas as pd
import time 
import datetime
from datetime import timedelta
import glob
import re
```

```python
url_geschaeft = 'https://www.gemeinderat-zuerich.ch/api/Geschaeft/suchen?orderBy=geschaeftsnummer&orderDir=desc&activePage=1&pageSize=20000'
response = requests.get(url_geschaeft)
vorstoesse = response.json()
vorstoesse = vorstoesse['Geschaefte']
print(len(vorstoesse))
```

#### Loop über alle Geschäfte, um Details herunterzuladen und in Datensatzform zu bringen

```python
data = list()
counter = len(vorstoesse)
for vorstoss in vorstoesse:
    url = 'https://www.gemeinderat-zuerich.ch/api/Geschaeft/details?gid='
    g_id = vorstoss['GeschaeftId']
    url = url+g_id
    response = requests.get(url)
    vordetails = response.json()
    if len(vordetails['EinreichendeGruppe']) == 0:
        vorstoss['gruppe'] = 'NaN'
        try:
            vorstoss['id_1'] = vordetails['Erstunterzeichnender']['Id']
            vorstoss['name_1'] = vordetails['Erstunterzeichnender']['Name']
            vorstoss['vorname_1'] = vordetails['Erstunterzeichnender']['Vorname']
            vorstoss['partei_1'] = vordetails['Erstunterzeichnender']['Partei']
            vorstoss['wahlkreis_1'] = vordetails['Erstunterzeichnender']['Wahlkreis']
            try:
                vorstoss['id_2'] = vordetails['ZweiterErstunterzeichnender']['Id']
                vorstoss['name_2'] = vordetails['ZweiterErstunterzeichnender']['Name']
                vorstoss['vorname_2'] = vordetails['ZweiterErstunterzeichnender']['Vorname']
                vorstoss['partei_2'] = vordetails['ZweiterErstunterzeichnender']['Partei']
                vorstoss['wahlkreis_2'] = vordetails['ZweiterErstunterzeichnender']['Wahlkreis']
            except:
                vorstoss['id_2'] = 'NaN'
                vorstoss['name_2'] = 'NaN'
                vorstoss['vorname_2'] = 'NaN'
                vorstoss['partei_2'] = 'NaN'
                vorstoss['wahlkreis_2'] = 'NaN'
        except:
            vorstoss['id_1'] = 'NaN'
            vorstoss['name_1'] = 'NaN'
            vorstoss['vorname_1'] = 'NaN'
            vorstoss['partei_1'] ='NaN'
            vorstoss['wahlkreis_1'] ='NaN'
            vorstoss['id_2'] = 'NaN'
            vorstoss['name_2'] = 'NaN'
            vorstoss['vorname_2'] = 'NaN'
            vorstoss['partei_2'] = 'NaN'
            vorstoss['wahlkreis_2'] = 'NaN'
    else:
        vorstoss['gruppe'] = vordetails['EinreichendeGruppe']
    counter -= 1
    print(counter)
    data.append(vorstoss.copy())
```

```python
df = pd.DataFrame(data)
df['Geschaeftstitel'] = df['Geschaeftstitel'].str.replace('\r',' ') #gab ein Fehler beim abspeichern. Einige Geschäfte haben offenbar ein \r im Titel
df.to_csv('gemeinderat_stand_'+today+'.csv', encoding='utf-8', index=False)
len(df)
```

```python
df['names1'] = df['vorname_1'] + " " + df['name_1']
df['names2'] = df['vorname_2'] + " " + df['name_2']
```

```python
def get_partners(name, year):
    polid = str(list(set(df['id_1'][df['names1']==name]))[0])
    polparty = str(list(set(df['partei_1'][df['names1']==name]))[0])
    
    subset = df[(df['id_1']==polid) & (df['Geschaeftsjahr']>=year) | (df['id_2']==polid) & (df['Geschaeftsjahr']>=year)]
    print(name + ' (' + polparty + ')' + ' hat ' + str(len(subset)) + ' Vorstösse eingereicht.')
    
    print('*****\n')
    print('Zusammen mit:\n')    

    partnerlist = pd.DataFrame(subset.names1.values.tolist() + subset.names2.values.tolist())
    partnerlist = partnerlist[(partnerlist[0]!='NaN NaN') & (partnerlist[0]!=name)]
    
    print(partnerlist[0].value_counts())
    print('\n****\n')
    print('So oft waren andere Parteien involviert\n')
    partnerparty = pd.DataFrame(subset.partei_1.values.tolist() + subset.partei_2.values.tolist())
    partnerparty = partnerparty[(partnerparty[0]!='NaN') & (partnerparty[0]!=polparty)]    
    print(partnerparty[0].value_counts())
    print('\nUnd in Prozent:')
    print(partnerparty[0].value_counts('normalize'))

    
    print('\n******\n')
    print('Das waren die Vorstossarten\n')
    art = pd.DataFrame(subset.Geschaeftsart.values.tolist())
    print(art[0].value_counts())
    print('\nUnd in Prozent:')
    print(art[0].value_counts('normalize'))
```
